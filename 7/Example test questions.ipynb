# Question 4

a summation from $i=1$ up to 40 $\sum_{i=1}^{40} e^i$ this means more less this: $e^1+e^2+...+e^{40}$  
a sequence multiplication of $\Pi_{i=1}^{40} e^i$ this means more less this:  $e^1*e^2*...*e^{40}$  

```c
double fun4(double a, double b, double c)
{
    double sum = 0;// and with 1
    for(int i=1; i<=40; i++)
    {
        sum += a * exp(1.0/(i+1.0)) + 1.0 / b;
    }
    return sum;
}
```

# Question 5

```c
double fun5(int n)
{
    double sum = 0;
    
    double x0 = 2.0;
    double xi;
    for(int i=1; i<=n; ++i)
    {
        xi = -0.5 * x0 * sqrt(fabs(x0));
        x0 = xi;
        sum += pow(-1, i+1) * 1.0 / xi;
    }
    
    return exp(sum);
}
```
